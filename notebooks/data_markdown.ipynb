```python
data_uniq_q = pd.read_csv("data/data_uiq_q.csv")
```

```python
data_uniq_q
```

```python
for_del = data_uniq_q[data_uniq_q.Answer.str.lower().str.contains('пользователь wb')]

for_drop = ['цена', 'пришло', 'где', 'долго', 'доставка', 'ожида', 'получ', 'есть ли', "гаранти"]
```

```python
data_uniq_q = data_uniq_q.drop(for_del.index)
```

```python
patterns_in_ques_neg = ["когда","вернуть", "сдать", "возврат", "деньги","акци", "наличи", "поступ",\
                     "отказ", "отмен", "ошиб", "поставк", "задерж", "купить", "скидк", "отдельно", "склад", "продаж"]

patterns_in_ans_neg = ["wb", "wildberries", "ситуаци", "обратитесь"]

filtered_dfs = []
for i in patterns_in_ques_neg:
    filtered_dfs.append(data_uniq_q[data_uniq_q['Question'].str.lower().str.contains(i)])

for i in patterns_in_ans_neg:
    filtered_dfs.append(data_uniq_q[data_uniq_q['Answer'].str.lower().str.contains(i)]) 

negative_data = pd.concat(filtered_dfs)
negative_data = negative_data.drop_duplicates()
```

```python
for i in for_drop:
    negative_data = negative_data.drop(negative_data[negative_data['Question'].str.lower().str.contains(i)].index)
```
